### Gauss Rank
Наш датасет содержит несколько фич без четкого распределения.
Попробуем нормализовать

```python
analyze_features = strange_features.copy()
df_base[analyze_features].hist(figsize=[10, 10], bins=100)
plt.show()
```

```python
def toGauss(base, train, validation, column):
    quantile = QuantileTransformer(random_state=42, output_distribution='normal')
    base_feature = base[column].values.reshape((len(base[column]),1))
    train_feature = train[column].values.reshape((len(train[column]),1))
    validation_feature = train[column].values.reshape((len(validation[column]),1))

    quantile.fit(base_feature)
    base_feature_normal = quantile.transform(base_feature)
    train_feature_normal = quantile.transform(train_feature)
    validation_feature_normal = quantile.transform(validation_feature)

    tbase = base.copy().reset_index()
    ttrain = train.copy().reset_index()
    tvalidation = validation.copy().reset_index()

    tbase[column+"_normal"] = pd.Series(base_feature_normal.reshape((len(base_feature_normal), 1)[0]))
    ttrain[column+"_normal"] = pd.Series(train_feature_normal.reshape((len(train_feature_normal), 1)[0]))
    tvalidation[column+"_normal"] = pd.Series(validation_feature_normal.reshape((len(validation_feature_normal), 1)[0]))
    
    tbase.set_index("Id", inplace=True)
    ttrain.set_index("Id", inplace=True)
    tvalidation.set_index("Id", inplace=True)

    
    return tbase, ttrain, tvalidation
```

```python
for i in analyze_features:
    df_base, df_train, df_validation = toGauss(df_base, df_train, df_validation, i)
```

```python
check_resullts = [i + "_normal" for i in analyze_features] + analyze_features
df_base[check_resullts].hist(bins=100, figsize=(15, 9))
plt.show()
```

## Проверим бейзлайн

```python
def checkHeadonModel(base, train, k_value=5):
    dims = base.shape[1]
    n_cells = 50
    # how many cells
    quantizer = faiss.IndexFlatL2(dims)
    idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
    
    idx_l2.train(np.ascontiguousarray(base.values).astype('float32'))
    idx_l2.add(np.ascontiguousarray(base.values).astype('float32'))
    
    base_index = {k: v for k, v in enumerate(base.index.to_list())}

    headon_solution_df_train = train.copy()
    targets = headon_solution_df_train["Target"]
    headon_solution_df_train.drop("Target", axis=1, inplace=True)
    
    k=k_value
    vecs, idx = idx_l2.search(np.ascontiguousarray(headon_solution_df_train.values).astype('float32'), k)
    
    acc = 0
    for target, el in zip(targets.values.tolist(), idx.tolist()):
        acc += int(target in [base_index[r] for r in el])
    
    print(f"A{k_value}Score:", 100 * acc / len(idx), "%")
```

```python
%%time
checkHeadonModel(df_base, df_train)
```

## Scalers
Попробуем для начала преобразовать данные и проверить, улучшится ли скор поиск по K__VALUE = 50 с разными Faiss индексами

```python
index_scalers_stats = pd.DataFrame(columns=["index", "scaler", "score"])
index_scalers_stats
```

```python
def buildA5Score(df_base, df_train, short=True):
    dims = df_base.shape[1]
    n_cells = 50
    # how many cells
    quantizer = faiss.IndexFlatL2(dims)
    idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
    idx_l2.train(np.ascontiguousarray(df_base.values).astype('float32'))
    idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32'))
    
    base_index = {k: v for k, v in enumerate(df_base.index.to_list())}
    #del df_base
    
    headon_solution_df_train = df_train.copy()
    targets = headon_solution_df_train["Target"]
    headon_solution_df_train.drop("Target", axis=1, inplace=True)

    k=50
    vecs, idx = idx_l2.search(np.ascontiguousarray(headon_solution_df_train.values).astype('float32'), k)
    
    acc = 0
    for target, el in zip(targets.values.tolist(), idx.tolist()):
        acc += int(target in [base_index[r] for r in el])
    
    return 100 * acc / len(idx)
```

### StandardScaler

```python
def checkStandardScaler(base, train):
    scaler = StandardScaler()
    scaler.fit(base)

    tbase = scaler.transform(base)
    tbase = pd.DataFrame(tbase, columns=base.columns)
    tbase["Id"] = base.index
    tbase = tbase.set_index(["Id"])
    
    ttrain = scaler.transform(train[base.columns])

    ttrain = pd.DataFrame(ttrain, columns=base.columns)
    ttrain["Id"] = train.index
    ttrain = ttrain.set_index(["Id"])
    ttrain["Target"] = train["Target"]

    return tbase, ttrain, scaler
```

```python
sscaler_base, sscaler_train, _ = checkStandardScaler(df_base, df_train)
```

```python
%%time
sscaler_score = buildA5Score(sscaler_base, sscaler_train)
print("A@50 Score on Standard Scaler :", buildA5Score(sscaler_base, sscaler_train))
```

#### IndexLSH

```python
def buildA5ScoreV2(df_base, df_train, short=True, k_value=50):
    dims = df_base.shape[1]
    n_cells = 100

    idx_l2 = faiss.IndexLSH(dims, 1024)
    if short:
        idx_l2.train(np.ascontiguousarray(df_base[:50000].values).astype('float32'))
        idx_l2.add(np.ascontiguousarray(df_base[:50000].values).astype('float32'))
    else:
        idx_l2.train(np.ascontiguousarray(df_base.values).astype('float32'))
        idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32'))
    
    base_index = {k: v for k, v in enumerate(df_base.index.to_list())}
    #del df_base
    
    headon_solution_df_train = df_train.copy()
    targets = headon_solution_df_train["Target"]
    headon_solution_df_train.drop("Target", axis=1, inplace=True)

    k=k_value
    vecs, idx = idx_l2.search(np.ascontiguousarray(headon_solution_df_train.values).astype('float32'), k)
    
    acc = 0
    for target, el in zip(targets.values.tolist(), idx.tolist()):
        acc += int(target in [base_index[r] for r in el])
    
    return 100 * acc / len(idx)
```

```python
def analysFeatureImportance(base, train):
    importance_features_df = pd.DataFrame(columns=["excluded_feature", "score"])
    original_column_list = base.columns.copy().to_list()
    
    for index in tqdm(range(len(original_column_list))):
        i = original_column_list[index]
        temp_columns = original_column_list.copy()
        temp_columns.remove(i)
        train_columns = temp_columns.copy()
        train_columns.append("Target")
        score = buildA5ScoreV2(base[temp_columns], train[train_columns])

        importance_features_df = pd.concat([importance_features_df, pd.DataFrame({"excluded_feature": str(i), "score": score}, index=[0])])

    return importance_features_df
```

### Обучение FAISS

#### A-N Score function

```python
def getAScore(targets, seached_idxs, base_index):
    defineActualBaseTargets = lambda x: base_index[x]
    
    start_score = time.time()
    acc = 0
    for target, el in zip(targets, seached_idxs):
        acc += int(np.isin(target, np.array([base_index[r] for r in el])))
    print("Score time:", time.time() - start_score)
    return 100 * acc / len(seached_idxs)
```

### Научим FAISS плохому

```python
def checkRobustScaler(base, train, validation):
    scaler = RobustScaler()
    scaler.fit(base)

    tbase = scaler.transform(base)
    tbase = pd.DataFrame(tbase, columns=base.columns)
    tbase["Id"] = base.index
    tbase = tbase.set_index(["Id"])
    
    ttrain = scaler.transform(train[base.columns])
    ttrain = pd.DataFrame(ttrain, columns=base.columns)
    ttrain["Id"] = train.index
    ttrain = ttrain.set_index(["Id"])
    ttrain["Target"] = train["Target"]

    tvalidation = scaler.transform(validation[base.columns])
    tvalidation = pd.DataFrame(tvalidation, columns=base.columns)
    tvalidation["Id"] = tvalidation.index
    tvalidation = tvalidation.set_index(["Id"])

    return tbase, ttrain, tvalidation, scaler
```

```python
def train_index(df_base, df_train, short=True, k_value=50):
    start_train = time.time()
    dims = df_base.shape[1]

    idx_l2 = faiss.IndexLSH(dims, 1024)
    headon_solution_df_train = df_train.copy()

    df_base_sample = df_base.copy()

    if short:
        df_base_sample = df_base[:50000]
        headon_solution_df_train = headon_solution_df_train[headon_solution_df_train["Target"].isin(df_base_sample.index)]

        idx_l2.train(np.ascontiguousarray(df_base_sample.values).astype('float32'))
        idx_l2.add(np.ascontiguousarray(df_base_sample.values).astype('float32'))
    else:
        idx_l2.train(np.ascontiguousarray(df_base.values).astype('float32'))
        idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32'))

    #base_index = {k: v for k, v in enumerate(df_base.index.to_list())}
    base_index = np.array(df_base.index.to_list())

    print("Learning index for:", time.time() - start_train)
    return df_base_sample, headon_solution_df_train, idx_l2, base_index
    
```

```python
def searchByIndex(index, train, k_value):
    targets = train["Target"]
    train.drop("Target", axis=1, inplace=True)

    search_start = time.time()
    k=k_value
    
    vecs, idx = index.search(np.ascontiguousarray(train.values).astype('float32'), k)

    search_time = time.time() - search_start
    return vecs, idx, search_time
```

```python
def buildAKScoreV2For(df_base, df_train, short=True, k_value=50):
    base, train, index, base_index = train_index(df_base, df_train, short, k_value)
    targets = train["Target"]
    
    vecs, idx, search_time = searchByIndex(index, train, k_value)
    score = getAScore(targets.values.tolist(), idx.tolist(), base_index)

    searched_targets = []
    for i in idx:
        searched_targets.append([base_index[k] for k in i])
    
    return score, index, searched_targets, (k_value, search_time, score), base_index
```

Применим Robust Scaler к нашим данным

```python
rscaler_base, rscaler_train, rscaler_validation, _ = checkRobustScaler(df_base, df_train, df_validation)
```

А теперь найдем оптимальное количество соседей для выборки в 50 000 запискй

```python
%%time
results = []
for kvalue in tqdm(list(range(5, 100, 5)) + list(range(100, 1000, 100))):
    lsh_score, lsh_index, targets, stats, _ = buildAKScoreV2For(rscaler_base, rscaler_train, short=True, k_value=kvalue)
    results.append(stats)
```
